### Objective #2: I want to investigate which region is the most consistent across its countries.

Creating new dataframe that includes the region and country along with the other numeric variables.

```python
df_region = df.loc[:,["Region", "Happiness Score", "Economy (GDP per Capita)", "Family", "Health (Life Expectancy)", "Freedom", "Trust (Government Corruption)", "Generosity", "Dystopia Residual"]]
df_region
```

Grouping by region and looking at the standard deviation of each region since I want to investigate which region is the most consistent of the countries within it. Regions that have a higher standard deviation could be argued that there is less consistency among its countries, and regions that have a lower standard deviation could be argued that there is more consistency amoung its countries. 

```python
df_region.groupby("Region").std()
```

Creating heat maps to visualize any differences between the regions. I want to use heat maps for this because it's a quick way to visualize which regions have a overall higher standard deviation or lower. 

```python
sns.heatmap(df_region.groupby("Region").std())
plt.title("Standard Deviation of Variables for Each Region")
```

Since each variable is on it's own different scale, the colors of the heat map are skewed because it's causing the standard deviation of some variables to appear very dark in color because it's on the same heatmap as other variables with much larger scales and therefore much larger standard deviations. In order to counter this, each column will need to be normalized across it's own scale. Normalizing each variable to the same scale will help view differences more easily across all variables on the same graph.

Creating a new data frame to set the region as the index and also to get rid of any string columns. 

```python
df_region_index = df_region.set_index("Region")
df_region_index
```

Normalizing the data by taking each data point and subtracting the minimum from it, and dividing by the range (max value-min value). Now I can take the mean, min, max, and std of this normalized data frame and use heat maps with it. 

```python
df_norm_regions = (df_region_index - df_region_index.min())/(df_region_index.max() - df_region_index.min())
df_norm_regions
```

```python
df_norm_regions.groupby("Region").std()
```

It is now much easier to visually see the differences between regions using the color scale.

```python
sns.heatmap(df_norm_regions.groupby("Region").std())
plt.title("Standard Deviation of Normalized Variables for Each Region")
```

Looking at the normalized heat map, we can see that darker colors are associated with lower standard deviations across the data points, and therefore less variety in those scores. The region with the most dark colored standard deviation scores is the Australia and New Zealand region. North America is another region with very low standard deviation scores across the variables, but not as many low standard deviation scores as Australia and New Zealand. Therefore, Australia and New Zealand is the region that is the most consistent across the countries within it in the data set. This would make sense because the Australia and New Zealand region along with the North America region have much fewer countries within those regions than some of the other regions such as western Europe. 

### Objective #3: I want to investigate which regions are the overall best and worst places to live according to the data set.

I am going to use the df_norm data frame I created in the second objective because I want to make sure the data points from each column is weighted equally across all columns. Since the original data frame had different number scales for each column, each data entry is weighted differently. To utilize the average of all scores across all columns, each column needs to be weighted equally or else some data points, such as happiness score (since it's on a scale of 0-7), would be much more influential than other columns on the total score. 

```python
df_norm
```

I am using the mean of each variable for each region from the groupby() method. Utilizing the sum of all scores from countries in a region instead of the mean of all the countries in the region would not be taking into account that some regions have much more countries in this data set than others. Therefore using the mean of all the countries of the region is the most fair way since it is taking the variable averages across all the countries in a region.

```python
df_norm.groupby("Region").mean()
```

By summing together the averages of all variables for each region, we can get a sum of all the averages for each region across all the numeric variables. Regions with a higher sum of all the averages across their variables have a higher average score per country in the region. Regions with a lower sum of all the averages across their variables have a lower average score per country in the region. Therefore, the region with the highest sum score can be argued that it has the highest average scores across all its countries and is best region to live in. On the flip side, the region with the lowest sum score can be argued that they have the lowest average scores across their countries and is the worst region to live in.

```python
df_norm.groupby("Region").mean().sum("columns")
```

Using a bargraph for visualization the total sum score of the averages for each variable across the countries in each region is a quick and easy way to see which regions have a higher and lower sum becuase we get to see them side by side and relative to each other.

```python
df_norm_bargraph = df_norm.groupby("Region").mean().sum("columns")

sns.barplot(x = df_norm_bargraph.index, y = df_norm_bargraph)
plt.xticks(rotation = 90)
plt.xlabel("Region")
plt.ylabel("Sum")
plt.title("Sum of the Averages of each normalized variable for all Countries in the Region")
```

According to this method, Australia and New Zealand have the highest sum score and Sub-Saharan Africa has the lowest sum score. Therefore, the Australia and New Zealand region is the best region to live in and the Sub-Saharan Africa is the worst region to live in according to this analysis of the dataset.
